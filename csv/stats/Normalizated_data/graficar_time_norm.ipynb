```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
anger_stats = pd.read_csv ('Stats_anger.csv')
disgust_stats = pd.read_csv ('Stats_disgust.csv')
fear_stats = pd.read_csv ('Stats_fear.csv')
happiness_stats = pd.read_csv ('Stats_happiness.csv')
neutral_stats = pd.read_csv ('Stats_neutral.csv')
sadness_stats = pd.read_csv ('Stats_sadness.csv')
surprise_stats = pd.read_csv ('Stats_surprise.csv')
```

### First moment

```python
plt.scatter(anger_stats['Unnamed: 0'], anger_stats['M1'], color = 'red', label = "Angry")
plt.scatter(disgust_stats['Unnamed: 0'], disgust_stats['M1'], color = 'olive', label = "Disgust")
plt.scatter(fear_stats['Unnamed: 0'], fear_stats['M1'], color = 'green', label = "Fear")
plt.scatter(happiness_stats['Unnamed: 0'], happiness_stats['M1'], color = 'darkviolet', label = "Happiness")
plt.scatter(sadness_stats['Unnamed: 0'], sadness_stats['M1'], color = 'darkturquoise', label = "Sadness")
plt.scatter(surprise_stats['Unnamed: 0'], surprise_stats['M1'], color = 'gold', label = "Surprise")
plt.scatter(neutral_stats['Unnamed: 0'], neutral_stats['M1'], color = 'deeppink', label = "Neutral")

plt.axvline(x = 14, color = 'dimgray', label = 'Person 1', linestyle = 'solid', linewidth = '1')
plt.axvline(x = 29, color = 'dimgray', label = 'Person 2', linestyle = 'dashdot', linewidth = '1')
plt.axvline(x = 44, color = 'dimgray', label = 'Person 3', linestyle = 'dashed', linewidth = '1')
plt.axvline(x = 59, color = 'dimgray', label = 'Person 4', linestyle = 'dotted', linewidth = '1')

plt.legend (loc = (1.005,0))
plt.suptitle('First Moment', fontsize = 20)
plt.xlabel ("Audio sample", fontsize = 13)
plt.ylabel ("M1", fontsize = 13)

plt.savefig ("images\M1_graph.jpg", bbox_inches="tight", dpi = 300)
```
