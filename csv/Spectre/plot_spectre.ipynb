# Comparar sectro en frecuencia de las mismas frases con diferentes sentimientos

### Se importan librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Importar lista con nombres de archivos

```python
files_Anger = pd.read_csv("..\Anger\Anger.csv") 
files_Disgust = pd.read_csv("..\Disgust\Disgust.csv") 
files_Fear = pd.read_csv("..\Fear\Fear.csv") 
files_Neutral = pd.read_csv(r"..\Neutral\Neutral.csv") 
files_Happiness = pd.read_csv("..\Happiness\Happiness.csv")
files_Sadness = pd.read_csv("..\Sadness\Sadness.csv") 
files_Surprise = pd.read_csv("..\Surprise\Surprise.csv") 
```

### Ciclo para graficar espectro de frases en diferentes sentimientos

```python
# Variables de la seal
#fs = 44100 #Frecuencia de la seal en Hz
fs = 41400
p = 1/fs   #Periodo de la seal en s


for a in range(3):

    #Anger
    name_doc = files_Anger._get_value (a, 'Name')
    data_Anger = pd.read_csv ("..\Anger\\" + name_doc, header = None)
    data_Anger = data_Anger.values.flatten()
    data_Anger = data_Anger - np.mean (data_Anger)
    A = np.fft.fft (data_Anger)
    Spectre_A = np.fft.fftshift (np.abs (A))
    w_A = np.linspace (-1, 1, len(data_Anger)) * (1/p) * (0.5)
    print(name_doc)

    #Disgust
    name_doc = files_Disgust._get_value (a, 'Name')
    data_Disgust = pd.read_csv ("..\Disgust\\" + name_doc, header = None)
    data_Disgust = data_Disgust.values.flatten()
    data_Disgust = data_Disgust - np.mean (data_Disgust)
    D = np.fft.fft (data_Disgust)
    Spectre_D = np.fft.fftshift (np.abs (D))
    w_D = np.linspace (-1, 1, len(data_Disgust)) * (1/p) * (0.5)
    print(name_doc)

    #Fear
    name_doc = files_Fear._get_value (a, 'Name')
    data_Fear = pd.read_csv ("..\Fear\\" + name_doc, header = None)
    data_Fear = data_Fear.values.flatten()
    data_Fear = data_Fear - np.mean (data_Fear)
    F = np.fft.fft (data_Fear)
    Spectre_F = np.fft.fftshift (np.abs (F))
    w_F = np.linspace (-1, 1, len(data_Fear)) * (1/p) * (0.5)
    print(name_doc)

    #Happiness
    name_doc = files_Happiness._get_value (a, 'Name')
    data_Happiness = pd.read_csv ("..\Happiness\\" + name_doc, header = None)
    data_Happiness = data_Happiness.values.flatten()
    data_Happiness = data_Happiness - np.mean (data_Happiness)
    H = np.fft.fft (data_Happiness)
    Spectre_H = np.fft.fftshift (np.abs (H))
    w_H = np.linspace (-1, 1, len(data_Happiness)) * (1/p) * (0.5)
    print(name_doc)

    #Neutral
    name_doc = files_Neutral._get_value (a, 'Name')
    data_Neutral = pd.read_csv (r"..\Neutral\\" + name_doc, header = None)
    data_Neutral = data_Neutral.values.flatten()
    data_Neutral = data_Neutral - np.mean (data_Neutral)
    N = np.fft.fft (data_Neutral)
    Spectre_N = np.fft.fftshift (np.abs (N))
    w_N = np.linspace (-1, 1, len(data_Neutral)) * (1/p) * (0.5)
    print(name_doc)

    #Sadness
    name_doc = files_Sadness._get_value (a, 'Name')
    data_Sadness = pd.read_csv ("..\Sadness\\" + name_doc, header = None)
    data_Sadness = data_Sadness.values.flatten()
    data_Sadness = data_Sadness - np.mean (data_Sadness)
    Sa = np.fft.fft (data_Sadness)
    Spectre_Sa = np.fft.fftshift (np.abs (Sa))
    w_Sa = np.linspace (-1, 1, len(data_Sadness)) * (1/p) * (0.5)
    print(name_doc)

    #Surprise
    name_doc = files_Surprise._get_value (a, 'Name')
    data_Surprise = pd.read_csv ("..\Surprise\\" + name_doc, header = None)
    data_Surprise = data_Surprise.values.flatten()
    data_Surprise = data_Surprise - np.mean (data_Surprise)
    Su = np.fft.fft (data_Surprise)
    Spectre_Su = np.fft.fftshift (np.abs (Su))
    w_Su = np.linspace (-1, 1, len(data_Surprise)) * (1/p) * (0.5)
    print(name_doc)

    plt.figure(a)
    plt.plot(w_A, Spectre_A, color = 'red', label = "Angry")
    plt.plot(w_D, Spectre_D, color = 'olive', label = "Disgust")
    plt.plot(w_F, Spectre_F, color = 'green', label = "Fear")
    plt.plot(w_H, Spectre_H, color = 'darkviolet', label = "Happiness")
    plt.plot(w_Sa, Spectre_Sa, color = 'darkturquoise', label = "Sadness")
    plt.plot(w_Su, Spectre_Su, color = 'gold', label = "Surprise")
    plt.plot(w_N, Spectre_N, color = 'deeppink', label = "Neutral")
    plt.xlim ([0,1000])
    plt.suptitle('Phrase ' + str (a+1) + ', Subject 1' , fontsize = 20)
    plt.legend (loc = (1.005,0))
    #plt.suptitle('Frequency spectrum')
    plt.xlabel ("Frequency (Hz)", fontsize = 13)
    plt.ylabel ("Amplitude", fontsize = 13)
    plt.grid()
    plt.show()
```
