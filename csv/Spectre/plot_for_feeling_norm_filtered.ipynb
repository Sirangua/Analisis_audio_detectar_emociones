# Se grafican todos los espectros filtrados de un sentimiento para ver si coinciden en algo

### Se importan librerias


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import scipy.stats as st
scaler = MinMaxScaler()
```

### Importar lista con nombres de archivos

```python
files_Anger = pd.read_csv("..\Anger\Anger.csv") 
files_Disgust = pd.read_csv("..\Disgust\Disgust.csv") 
files_Fear = pd.read_csv("..\Fear\Fear.csv") 
files_Neutral = pd.read_csv(r"..\Neutral\Neutral.csv") 
files_Happiness = pd.read_csv("..\Happiness\Happiness.csv")
files_Sadness = pd.read_csv("..\Sadness\Sadness.csv") 
files_Surprise = pd.read_csv("..\Surprise\Surprise.csv") 
```

```python
def gkern(kernlen=21, nsig=3):

    x = np.linspace(-nsig, nsig, kernlen+1)
    kern1d = np.diff(st.norm.cdf(x))
    return kern1d

KGauss = gkern(50,3)*275
#print(KGauss)

KGauss = np.array ([ 0.1755814,   0.24796536,  0.34518915,  0.4736711,   0.64069341,  0.85423479,
  1.1226849 ,  1.45442772 , 1.85729138 , 2.33787659,  2.90079316 , 3.54785275
,  4.27728371,  5.0830476  , 5.95434434 , 6.87539046,  7.82554228 , 8.77981109
,  9.70978374, 10.58492049 ,11.37415916 ,12.04771432, 12.57892942 ,12.94602256
, 13.13356716, 13.13356716 ,12.94602256 ,12.57892942, 12.04771432 ,11.37415916
, 10.58492049,  9.70978374 , 8.77981109 , 7.82554228,  6.87539046 , 5.95434434
,  5.0830476 ,  4.27728371 , 3.54785275 , 2.90079316,  2.33787659 , 1.85729138
,  1.45442772,  1.1226849  , 0.85423479 , 0.64069341,  0.4736711  , 0.34518915
,  0.24796536,  0.1755814 ])


 
```

### Se crea una grafica con todos los espectros de un sentimiento

Enojo


```python
# Variables de la señal
#fs = 44100 #Frecuencia de la señal en Hz
fs = 41400
p = 1/fs   #Periodo de la señal en s

plt.figure("Anger")
plt.xlim ([0,1000])
plt.suptitle('Spectres for anger' , fontsize = 20)
plt.xlabel ("Frequency (Hz)", fontsize = 13)
plt.ylabel ("Amplitude", fontsize = 13)
plt.grid()

for a in range(1):

    #Anger
    name_doc = files_Anger._get_value (a, 'Name')
    data_Anger = pd.read_csv ("..\Anger\\" + name_doc, header = None)
    data_Anger = data_Anger.values.flatten()
    data_Anger = data_Anger - np.mean (data_Anger)
    A = np.convolve (np.fft.fft (data_Anger), KGauss, mode = 'same').reshape(-1,1)
    Spectre_A = scaler.fit_transform(np.fft.fftshift (np.abs (A)))

    w_A = np.linspace (-1, 1, len(data_Anger)) * (1/p) * (0.5)

    plt.plot(w_A, Spectre_A)
    
plt.show("Anger")
```
