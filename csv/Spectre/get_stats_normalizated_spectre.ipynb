# Extraer variables estadisticas del espectro normalizdo de cada audio

### Se importan librerias

```python
import pandas as pd
from scipy import stats
import math
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

### Importar lista con nombres de archivos

```python
files_Anger = pd.read_csv("..\Anger\Anger.csv") 
files_Disgust = pd.read_csv("..\Disgust\Disgust.csv") 
files_Fear = pd.read_csv("..\Fear\Fear.csv") 
files_Neutral = pd.read_csv(r"..\Neutral\Neutral.csv") 
files_Happiness = pd.read_csv("..\Happiness\Happiness.csv")
files_Sadness = pd.read_csv("..\Sadness\Sadness.csv") 
files_Surprise = pd.read_csv("..\Surprise\Surprise.csv") 
```

### Se crea un diccionario para almacenar las variables de cada archivo

```python
data_instance_a = {'M1': [],
                 'M2': [],
                 'M3': [],
                 'M4': [],
                 'C2': [],
                 'C3': [],
                 'C4': [],
                 'Skewness': [],
                 'Kurtosis': [],
                 'Mean': [],
                 'Absolute_mean': [],
                 'Peak_value': [],
                 'SRV': [],
                 'RMS': [],
                 'Crest_factor': [],
                 'Shape_factor': []}
                 
data_instance_d = {'M1': [],
                 'M2': [],
                 'M3': [],
                 'M4': [],
                 'C2': [],
                 'C3': [],
                 'C4': [],
                 'Skewness': [],
                 'Kurtosis': [],
                 'Mean': [],
                 'Absolute_mean': [],
                 'Peak_value': [],
                 'SRV': [],
                 'RMS': [],
                 'Crest_factor': [],
                 'Shape_factor': []}

data_instance_f = {'M1': [],
                 'M2': [],
                 'M3': [],
                 'M4': [],
                 'C2': [],
                 'C3': [],
                 'C4': [],
                 'Skewness': [],
                 'Kurtosis': [],
                 'Mean': [],
                 'Absolute_mean': [],
                 'Peak_value': [],
                 'SRV': [],
                 'RMS': [],
                 'Crest_factor': [],
                 'Shape_factor': []}

data_instance_n = {'M1': [],
                 'M2': [],
                 'M3': [],
                 'M4': [],
                 'C2': [],
                 'C3': [],
                 'C4': [],
                 'Skewness': [],
                 'Kurtosis': [],
                 'Mean': [],
                 'Absolute_mean': [],
                 'Peak_value': [],
                 'SRV': [],
                 'RMS': [],
                 'Crest_factor': [],
                 'Shape_factor': []}

data_instance_h = {'M1': [],
                 'M2': [],
                 'M3': [],
                 'M4': [],
                 'C2': [],
                 'C3': [],
                 'C4': [],
                 'Skewness': [],
                 'Kurtosis': [],
                 'Mean': [],
                 'Absolute_mean': [],
                 'Peak_value': [],
                 'SRV': [],
                 'RMS': [],
                 'Crest_factor': [],
                 'Shape_factor': []}

data_instance_sa = {'M1': [],
                 'M2': [],
                 'M3': [],
                 'M4': [],
                 'C2': [],
                 'C3': [],
                 'C4': [],
                 'Skewness': [],
                 'Kurtosis': [],
                 'Mean': [],
                 'Absolute_mean': [],
                 'Peak_value': [],
                 'SRV': [],
                 'RMS': [],
                 'Crest_factor': [],
                 'Shape_factor': []}

data_instance_su = {'M1': [],
                 'M2': [],
                 'M3': [],
                 'M4': [],
                 'C2': [],
                 'C3': [],
                 'C4': [],
                 'Skewness': [],
                 'Kurtosis': [],
                 'Mean': [],
                 'Absolute_mean': [],
                 'Peak_value': [],
                 'SRV': [],
                 'RMS': [],
                 'Crest_factor': [],
                 'Shape_factor': []}
```

### Ciclo para obtener las variables estadisticas de cada sentimiento

```python
################################## Anger ##################################
for a in range(60):

    #Anger
    name_doc = files_Anger._get_value (a, 'Name')
    data_Anger = pd.read_csv ("..\Anger\\" + name_doc, header = None)
    data_Anger = data_Anger.values.flatten()
    data_Anger = data_Anger - np.mean (data_Anger)
    A = np.fft.fft (data_Anger).reshape(-1,1)
    Spectre_A = scaler.fit_transform(np.fft.fftshift (np.abs (A)))

    data = pd.DataFrame(Spectre_A)

    ################# Get statistical features #################
    m1 = float (stats.moment (data, moment = 1)) #First moment 
    m2 = float (stats.moment (data, moment = 2)) #Second moment
    m3 = float (stats.moment (data, moment = 3))#Third moment
    m4 = float (stats.moment (data, moment = 4)) #Fourth moment
    sc = float (m2 - math.sqrt (pow (m1,2))) #Second cumulant
    tc = float (m3 - (3*m1*m2) - (2*pow (m1,3))) #Third cumulant
    fc = float (m4 + (3*m3*m1) - (3*pow (m2,2)) + (12*m2*pow (m1,2)) - (6*pow (m1,4))) #Fourth cumulant
    skew = float (stats.skew (data)) #Skewness
    kurt = float (stats.kurtosis (data)) #Kurtosis
    mean = float(data.mean())# Mean
    am = abs(float(data.mean())) #Absolute mean
    pv = 0.5*(float(data.max()) - float(data.min())) #Peak value
    da = abs(data)
    dsa = da**2
    srv = float(pow((dsa.sum()/int(dsa.count())),2)) #Square root value
    d2 = data**2
    rms = math.sqrt (float(d2.sum())/int(d2.count())) #Root mean square
    cf = pv/rms #Crest factor
    sf = rms/am #Shape factor

    ################# Append features to a dictionary #################
    data_instance_a ['M1'].append(m1)
    data_instance_a ['M2'].append(m2)
    data_instance_a ['M3'].append(m3)
    data_instance_a ['M4'].append(m4)
    data_instance_a ['C2'].append(sc)
    data_instance_a ['C3'].append(tc)
    data_instance_a ['C4'].append(fc)
    data_instance_a ['Skewness'].append(skew)
    data_instance_a ['Kurtosis'].append(kurt)
    data_instance_a ['Mean'].append(mean)
    data_instance_a ['Absolute_mean'].append(am)
    data_instance_a ['Peak_value'].append(pv)
    data_instance_a ['SRV'].append(srv)
    data_instance_a ['RMS' ].append(rms)
    data_instance_a ['Crest_factor'].append(cf)
    data_instance_a ['Shape_factor'].append(sf)

    del data
data_stats = pd.DataFrame(data=data_instance_a)
data_stats.to_csv ("Graphs\Anger_norm_fur.csv")

################################## Disgust ##################################
for a in range(60):

    #Disgust
    name_doc = files_Disgust._get_value (a, 'Name')
    data_Disgust = pd.read_csv ("..\Disgust\\" + name_doc, header = None)
    data_Disgust = data_Disgust.values.flatten()
    data_Disgust = data_Disgust - np.mean (data_Disgust)
    D = np.fft.fft (data_Disgust).reshape(-1,1)
    Spectre_D = scaler.fit_transform(np.fft.fftshift (np.abs (D)))

    data = pd.DataFrame(Spectre_D)

    ################# Get statistical features #################
    m1 = float (stats.moment (data, moment = 1)) #First moment 
    m2 = float (stats.moment (data, moment = 2)) #Second moment
    m3 = float (stats.moment (data, moment = 3))#Third moment
    m4 = float (stats.moment (data, moment = 4)) #Fourth moment
    sc = float (m2 - math.sqrt (pow (m1,2))) #Second cumulant
    tc = float (m3 - (3*m1*m2) - (2*pow (m1,3))) #Third cumulant
    fc = float (m4 + (3*m3*m1) - (3*pow (m2,2)) + (12*m2*pow (m1,2)) - (6*pow (m1,4))) #Fourth cumulant
    skew = float (stats.skew (data)) #Skewness
    kurt = float (stats.kurtosis (data)) #Kurtosis
    mean = float(data.mean())# Mean
    am = abs(float(data.mean())) #Absolute mean
    pv = 0.5*(float(data.max()) - float(data.min())) #Peak value
    da = abs(data)
    dsa = da**2
    srv = float(pow((dsa.sum()/int(dsa.count())),2)) #Square root value
    d2 = data**2
    rms = math.sqrt (float(d2.sum())/int(d2.count())) #Root mean square
    cf = pv/rms #Crest factor
    sf = rms/am #Shape factor

    ################# Append features to a dictionary #################
    data_instance_d ['M1'].append(m1)
    data_instance_d ['M2'].append(m2)
    data_instance_d ['M3'].append(m3)
    data_instance_d ['M4'].append(m4)
    data_instance_d ['C2'].append(sc)
    data_instance_d ['C3'].append(tc)
    data_instance_d ['C4'].append(fc)
    data_instance_d ['Skewness'].append(skew)
    data_instance_d ['Kurtosis'].append(kurt)
    data_instance_d ['Mean'].append(mean)
    data_instance_d ['Absolute_mean'].append(am)
    data_instance_d ['Peak_value'].append(pv)
    data_instance_d ['SRV'].append(srv)
    data_instance_d ['RMS' ].append(rms)
    data_instance_d ['Crest_factor'].append(cf)
    data_instance_d ['Shape_factor'].append(sf)

    del data
data_stats = pd.DataFrame(data=data_instance_d)
data_stats.to_csv ("Graphs\Disgust_norm_fur.csv")
   

################################## Fear ##################################
for a in range(60):

    #Fear
    name_doc = files_Fear._get_value (a, 'Name')
    data_Fear = pd.read_csv ("..\Fear\\" + name_doc, header = None)
    data_Fear = data_Fear.values.flatten()
    data_Fear = data_Fear - np.mean (data_Fear)
    F = np.fft.fft (data_Fear).reshape(-1,1)
    Spectre_F = scaler.fit_transform(np.fft.fftshift (np.abs (F)))

    data = pd.DataFrame(Spectre_F)

    ################# Get statistical features #################
    m1 = float (stats.moment (data, moment = 1)) #First moment 
    m2 = float (stats.moment (data, moment = 2)) #Second moment
    m3 = float (stats.moment (data, moment = 3))#Third moment
    m4 = float (stats.moment (data, moment = 4)) #Fourth moment
    sc = float (m2 - math.sqrt (pow (m1,2))) #Second cumulant
    tc = float (m3 - (3*m1*m2) - (2*pow (m1,3))) #Third cumulant
    fc = float (m4 + (3*m3*m1) - (3*pow (m2,2)) + (12*m2*pow (m1,2)) - (6*pow (m1,4))) #Fourth cumulant
    skew = float (stats.skew (data)) #Skewness
    kurt = float (stats.kurtosis (data)) #Kurtosis
    mean = float(data.mean())# Mean
    am = abs(float(data.mean())) #Absolute mean
    pv = 0.5*(float(data.max()) - float(data.min())) #Peak value
    da = abs(data)
    dsa = da**2
    srv = float(pow((dsa.sum()/int(dsa.count())),2)) #Square root value
    d2 = data**2
    rms = math.sqrt (float(d2.sum())/int(d2.count())) #Root mean square
    cf = pv/rms #Crest factor
    sf = rms/am #Shape factor

    ################# Append features to a dictionary #################
    data_instance_f ['M1'].append(m1)
    data_instance_f ['M2'].append(m2)
    data_instance_f ['M3'].append(m3)
    data_instance_f ['M4'].append(m4)
    data_instance_f ['C2'].append(sc)
    data_instance_f ['C3'].append(tc)
    data_instance_f ['C4'].append(fc)
    data_instance_f ['Skewness'].append(skew)
    data_instance_f ['Kurtosis'].append(kurt)
    data_instance_f ['Mean'].append(mean)
    data_instance_f ['Absolute_mean'].append(am)
    data_instance_f ['Peak_value'].append(pv)
    data_instance_f ['SRV'].append(srv)
    data_instance_f ['RMS' ].append(rms)
    data_instance_f ['Crest_factor'].append(cf)
    data_instance_f ['Shape_factor'].append(sf)

    del data
data_stats = pd.DataFrame(data=data_instance_f)
data_stats.to_csv ("Graphs\Fear_norm_fur.csv")


################################## Happiness ##################################
for a in range(60):

    #Happiness
    name_doc = files_Happiness._get_value (a, 'Name')
    data_Happiness = pd.read_csv ("..\Happiness\\" + name_doc, header = None)
    data_Happiness = data_Happiness.values.flatten()
    data_Happiness = data_Happiness - np.mean (data_Happiness)
    H = np.fft.fft (data_Happiness).reshape(-1,1)
    Spectre_H = scaler.fit_transform(np.fft.fftshift (np.abs (H)))

    data = pd.DataFrame(Spectre_H)

    ################# Get statistical features #################
    m1 = float (stats.moment (data, moment = 1)) #First moment 
    m2 = float (stats.moment (data, moment = 2)) #Second moment
    m3 = float (stats.moment (data, moment = 3))#Third moment
    m4 = float (stats.moment (data, moment = 4)) #Fourth moment
    sc = float (m2 - math.sqrt (pow (m1,2))) #Second cumulant
    tc = float (m3 - (3*m1*m2) - (2*pow (m1,3))) #Third cumulant
    fc = float (m4 + (3*m3*m1) - (3*pow (m2,2)) + (12*m2*pow (m1,2)) - (6*pow (m1,4))) #Fourth cumulant
    skew = float (stats.skew (data)) #Skewness
    kurt = float (stats.kurtosis (data)) #Kurtosis
    mean = float(data.mean())# Mean
    am = abs(float(data.mean())) #Absolute mean
    pv = 0.5*(float(data.max()) - float(data.min())) #Peak value
    da = abs(data)
    dsa = da**2
    srv = float(pow((dsa.sum()/int(dsa.count())),2)) #Square root value
    d2 = data**2
    rms = math.sqrt (float(d2.sum())/int(d2.count())) #Root mean square
    cf = pv/rms #Crest factor
    sf = rms/am #Shape factor

    ################# Append features to a dictionary #################
    data_instance_h ['M1'].append(m1)
    data_instance_h ['M2'].append(m2)
    data_instance_h ['M3'].append(m3)
    data_instance_h ['M4'].append(m4)
    data_instance_h ['C2'].append(sc)
    data_instance_h ['C3'].append(tc)
    data_instance_h ['C4'].append(fc)
    data_instance_h ['Skewness'].append(skew)
    data_instance_h ['Kurtosis'].append(kurt)
    data_instance_h ['Mean'].append(mean)
    data_instance_h ['Absolute_mean'].append(am)
    data_instance_h ['Peak_value'].append(pv)
    data_instance_h ['SRV'].append(srv)
    data_instance_h ['RMS' ].append(rms)
    data_instance_h ['Crest_factor'].append(cf)
    data_instance_h ['Shape_factor'].append(sf)

    del data
data_stats = pd.DataFrame(data=data_instance_h)
data_stats.to_csv ("Graphs\Happiness_norm_fur.csv")



################################## Neutral ##################################
for a in range(60):

    #Neutral
    name_doc = files_Neutral._get_value (a, 'Name')
    data_Neutral = pd.read_csv (r"..\Neutral\\" + name_doc, header = None)
    data_Neutral = data_Neutral.values.flatten()
    data_Neutral = data_Neutral - np.mean (data_Neutral)
    N = np.fft.fft (data_Neutral).reshape(-1,1)
    Spectre_N = scaler.fit_transform(np.fft.fftshift (np.abs (N)))

    data = pd.DataFrame(Spectre_N)

    ################# Get statistical features #################
    m1 = float (stats.moment (data, moment = 1)) #First moment 
    m2 = float (stats.moment (data, moment = 2)) #Second moment
    m3 = float (stats.moment (data, moment = 3))#Third moment
    m4 = float (stats.moment (data, moment = 4)) #Fourth moment
    sc = float (m2 - math.sqrt (pow (m1,2))) #Second cumulant
    tc = float (m3 - (3*m1*m2) - (2*pow (m1,3))) #Third cumulant
    fc = float (m4 + (3*m3*m1) - (3*pow (m2,2)) + (12*m2*pow (m1,2)) - (6*pow (m1,4))) #Fourth cumulant
    skew = float (stats.skew (data)) #Skewness
    kurt = float (stats.kurtosis (data)) #Kurtosis
    mean = float(data.mean())# Mean
    am = abs(float(data.mean())) #Absolute mean
    pv = 0.5*(float(data.max()) - float(data.min())) #Peak value
    da = abs(data)
    dsa = da**2
    srv = float(pow((dsa.sum()/int(dsa.count())),2)) #Square root value
    d2 = data**2
    rms = math.sqrt (float(d2.sum())/int(d2.count())) #Root mean square
    cf = pv/rms #Crest factor
    sf = rms/am #Shape factor

    ################# Append features to a dictionary #################
    data_instance_n ['M1'].append(m1)
    data_instance_n ['M2'].append(m2)
    data_instance_n ['M3'].append(m3)
    data_instance_n ['M4'].append(m4)
    data_instance_n ['C2'].append(sc)
    data_instance_n ['C3'].append(tc)
    data_instance_n ['C4'].append(fc)
    data_instance_n ['Skewness'].append(skew)
    data_instance_n ['Kurtosis'].append(kurt)
    data_instance_n ['Mean'].append(mean)
    data_instance_n ['Absolute_mean'].append(am)
    data_instance_n ['Peak_value'].append(pv)
    data_instance_n ['SRV'].append(srv)
    data_instance_n ['RMS' ].append(rms)
    data_instance_n ['Crest_factor'].append(cf)
    data_instance_n ['Shape_factor'].append(sf)

    del data
data_stats = pd.DataFrame(data=data_instance_n)
data_stats.to_csv (r"Graphs\Neutral_norm_fur.csv")



################################## Sadness ##################################
for a in range(60):

    #Sadness
    name_doc = files_Sadness._get_value (a, 'Name')
    data_Sadness = pd.read_csv ("..\Sadness\\" + name_doc, header = None)
    data_Sadness = data_Sadness.values.flatten()
    data_Sadness = data_Sadness - np.mean (data_Sadness)
    Sa = np.fft.fft (data_Sadness).reshape(-1,1)
    Spectre_Sa = scaler.fit_transform(np.fft.fftshift (np.abs (Sa)))

    data = pd.DataFrame(Spectre_Sa)

    ################# Get statistical features #################
    m1 = float (stats.moment (data, moment = 1)) #First moment 
    m2 = float (stats.moment (data, moment = 2)) #Second moment
    m3 = float (stats.moment (data, moment = 3))#Third moment
    m4 = float (stats.moment (data, moment = 4)) #Fourth moment
    sc = float (m2 - math.sqrt (pow (m1,2))) #Second cumulant
    tc = float (m3 - (3*m1*m2) - (2*pow (m1,3))) #Third cumulant
    fc = float (m4 + (3*m3*m1) - (3*pow (m2,2)) + (12*m2*pow (m1,2)) - (6*pow (m1,4))) #Fourth cumulant
    skew = float (stats.skew (data)) #Skewness
    kurt = float (stats.kurtosis (data)) #Kurtosis
    mean = float(data.mean())# Mean
    am = abs(float(data.mean())) #Absolute mean
    pv = 0.5*(float(data.max()) - float(data.min())) #Peak value
    da = abs(data)
    dsa = da**2
    srv = float(pow((dsa.sum()/int(dsa.count())),2)) #Square root value
    d2 = data**2
    rms = math.sqrt (float(d2.sum())/int(d2.count())) #Root mean square
    cf = pv/rms #Crest factor
    sf = rms/am #Shape factor

    ################# Append features to a dictionary #################
    data_instance_sa ['M1'].append(m1)
    data_instance_sa ['M2'].append(m2)
    data_instance_sa ['M3'].append(m3)
    data_instance_sa ['M4'].append(m4)
    data_instance_sa ['C2'].append(sc)
    data_instance_sa ['C3'].append(tc)
    data_instance_sa ['C4'].append(fc)
    data_instance_sa ['Skewness'].append(skew)
    data_instance_sa ['Kurtosis'].append(kurt)
    data_instance_sa ['Mean'].append(mean)
    data_instance_sa ['Absolute_mean'].append(am)
    data_instance_sa ['Peak_value'].append(pv)
    data_instance_sa ['SRV'].append(srv)
    data_instance_sa ['RMS' ].append(rms)
    data_instance_sa ['Crest_factor'].append(cf)
    data_instance_sa ['Shape_factor'].append(sf)

    del data
data_stats = pd.DataFrame(data=data_instance_sa)
data_stats.to_csv ("Graphs\Sadness_norm_fur.csv")



################################## Surprise ##################################
for a in range(60):

    #Surprise
    name_doc = files_Surprise._get_value (a, 'Name')
    data_Surprise = pd.read_csv ("..\Surprise\\" + name_doc, header = None)
    data_Surprise = data_Surprise.values.flatten()
    data_Surprise = data_Surprise - np.mean (data_Surprise)
    Su = np.fft.fft (data_Surprise).reshape(-1,1)
    Spectre_Su = scaler.fit_transform(np.fft.fftshift (np.abs (Su)))

    data = pd.DataFrame(Spectre_Su)

    ################# Get statistical features #################
    m1 = float (stats.moment (data, moment = 1)) #First moment 
    m2 = float (stats.moment (data, moment = 2)) #Second moment
    m3 = float (stats.moment (data, moment = 3))#Third moment
    m4 = float (stats.moment (data, moment = 4)) #Fourth moment
    sc = float (m2 - math.sqrt (pow (m1,2))) #Second cumulant
    tc = float (m3 - (3*m1*m2) - (2*pow (m1,3))) #Third cumulant
    fc = float (m4 + (3*m3*m1) - (3*pow (m2,2)) + (12*m2*pow (m1,2)) - (6*pow (m1,4))) #Fourth cumulant
    skew = float (stats.skew (data)) #Skewness
    kurt = float (stats.kurtosis (data)) #Kurtosis
    mean = float(data.mean())# Mean
    am = abs(float(data.mean())) #Absolute mean
    pv = 0.5*(float(data.max()) - float(data.min())) #Peak value
    da = abs(data)
    dsa = da**2
    srv = float(pow((dsa.sum()/int(dsa.count())),2)) #Square root value
    d2 = data**2
    rms = math.sqrt (float(d2.sum())/int(d2.count())) #Root mean square
    cf = pv/rms #Crest factor
    sf = rms/am #Shape factor

    ################# Append features to a dictionary #################
    data_instance_su ['M1'].append(m1)
    data_instance_su ['M2'].append(m2)
    data_instance_su ['M3'].append(m3)
    data_instance_su ['M4'].append(m4)
    data_instance_su ['C2'].append(sc)
    data_instance_su ['C3'].append(tc)
    data_instance_su ['C4'].append(fc)
    data_instance_su ['Skewness'].append(skew)
    data_instance_su ['Kurtosis'].append(kurt)
    data_instance_su ['Mean'].append(mean)
    data_instance_su ['Absolute_mean'].append(am)
    data_instance_su ['Peak_value'].append(pv)
    data_instance_su ['SRV'].append(srv)
    data_instance_su ['RMS' ].append(rms)
    data_instance_su ['Crest_factor'].append(cf)
    data_instance_su ['Shape_factor'].append(sf)

    del data
data_stats = pd.DataFrame(data=data_instance_su)
data_stats.to_csv ("Graphs\Surprise_norm_fur.csv")
```
